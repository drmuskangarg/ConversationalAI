How will you find unusual words from the text. The words which are not in the corpus but there in the text.

```python
def unusual_words(text):
  text_vocab= set(w for w in text if w.isalpha())
  english_vocab = set(w for w in words.words())
  unusual_words= text_vocab.difference(english_vocab)
  return unusual_words
```

A wordlist is useful for solving word puzzles. How many words of six letters or more can you make from those shown? Each letter must be used once only and must contain center letter.

The alphabets are E, G, I, V, R, V, O, N, I


```python
freq_letters=nltk.FreqDist('egivrvoni')
Obligatory='r'
wordlist=words.words()
[w for w in wordlist if (len(w) >= 6 and Obligatory in w and nltk.FreqDist(w) <= freq_letters)]
```

Find how much content of your file is not the stopword?

```python
def fraction_not_stopword(text):
  content_non_stopword= [w for w in text if w not in stopwords.words(fileids='english')]
  return len(content_non_stopword)/len(text)
```

```python
fraction_not_stopword(gutenberg.words(fileids='austen-emma.txt'))
```

Conditional Frequency Distribution

It divides the text into groups and then find frequency distribution in each group.

We add a further condition before frequency distribution.

```python
cfd=nltk.ConditionalFreqDist(
    (fileid, name[-1])
    for fileid in names.fileids()
    for name in names.words(fileid)
)
```

```python
cfd.plot()
```

# WordNet Corpus: 

A semantically oriented dictionary.

It contains something more than meanings of words. It contains only English langauge words as of now (but a project is going on where Asian and European language words are also to be included in near future) with 155,287 english words and 117,659 synony sets.

1. Synsets or Synonym Sets

Synonym set returns different sets where each set corresponds to one particular meaning/ context of the word.



```python
wordnet.synsets('dog')
```

The word 'dog' have 8 different contexts in which it can be used in the text. 

Three part synset: (a.b.c)
a: Name of that context
b: Part of Speech of that context
c: The number in that Part of Speech

```python
wordnet.synset('dog.n.01').definition()
```

```python
wordnet.synset('dog.n.03').examples()
```

```python
wordnet.synset('dog.n.01').lemmas()
```

Note that this lemma is different than the ones that we have covered in Stemming and Lemmatization.

Here Lemma means the synonym that are used in different synsets.

Relationship between synsets

These synsets are arranges in the form of trees and there are some relationships in the tree and you can traverse that tree to find these relationships.

1. Hyponyms: It is the relationship between subordinates (all the synsets which share the same parent in a tree)

```python
wordnet.synset('car.n.01').hyponyms()
```

2. Hypernyms: The relationship between superordinates

```python
wordnet.synset('car.n.01').hypernyms()
```

```python
wordnet.synset('motor_vehicle.n.01').hypernyms()
```

3. Root Hypernyms



```python
wordnet.synset('car.n.01').root_hypernyms()
```

4. Part Meronyms

Parts or components of the synsets

```python
wordnet.synset('car.n.01').part_meronyms()
```

5. Substance: Parts or the components of the living agent

```python
wordnet.synset('tree.n.01').substance_meronyms()
```
